````python
import pandas as pd

# 1) Toplam satır sayısı
n_rows = len(train_raw)

# 2) Click ve Order oranları
click_rate  = train_raw["clicked"].mean()
order_rate  = train_raw["ordered"].mean()

# 3) Click -> Order dönüşüm oranı
# Yalnızca click edilen ürünler içinde order edilenlerin oranı
click_to_order_ratio = train_raw["ordered"].sum() / train_raw["clicked"].sum()

print(f"Toplam satır sayısı: {n_rows:,}")
print(f"Clicked oranı:  {click_rate:.4%}")
print(f"Ordered oranı:  {order_rate:.4%}")
print(f"Click → Order dönüşüm oranı: {click_to_order_ratio:.2%}")
````

📌 Bu çıktılar bize şunu söylüyor:

Click modeli zaten dengesiz (%3.5).

Order modeli çok daha zor (%0.038).

İki aşamalı modelleme (Click + Order|Click) bu yüzden şart.

````python
# Session bazlı özet
session_stats = (
    train_raw.groupby("session_id")
    .agg(
        n_items   = ("content_id_hashed", "count"),
        n_clicks  = ("clicked", "sum"),
        n_orders  = ("ordered", "sum"),
    )
    .reset_index()
)

print(session_stats.describe())

# Kaç oturum var?
print("Toplam session sayısı:", session_stats.shape[0])

# Click içeren oturum oranı
click_sessions = (session_stats["n_clicks"] > 0).mean()
print(f"Click içeren oturum oranı: {click_sessions:.2%}")

# Order içeren oturum oranı
order_sessions = (session_stats["n_orders"] > 0).mean()
print(f"Order içeren oturum oranı: {order_sessions:.2%}")

# Ortalama oturum başına click ve order sayısı
avg_clicks = session_stats["n_clicks"].mean()
avg_orders = session_stats["n_orders"].mean()
print(f"Oturum başına ortalama click sayısı: {avg_clicks:.2f}")
print(f"Oturum başına ortalama order sayısı: {avg_orders:.2f}")
````

1. Oturum Boyutu

Ortalama oturumda 127 ürün var (n_items.mean ≈ 127).

Ama dağılım çok geniş: minimum 25, maksimum 7607 ürün.
→ Demek ki bazı oturumlar çok kısa, bazıları aşırı uzun (muhtemelen popüler aramalarda).
→ Feature adayı: oturum uzunluğu (kaç ürün var).

````python
# User bazlı özet
user_stats = (
    train_raw.groupby("user_id_hashed")
    .agg(
        n_sessions = ("session_id", "nunique"),
        n_clicks   = ("clicked", "sum"),
        n_orders   = ("ordered", "sum"),
    )
    .reset_index()
)

print(user_stats.describe())

# Toplam kullanıcı sayısı
print("Toplam kullanıcı sayısı:", user_stats.shape[0])

# Order yapan kullanıcı oranı
order_users = (user_stats["n_orders"] > 0).mean()
print(f"En az 1 sipariş yapan kullanıcı oranı: {order_users:.2%}")

# Click yapan kullanıcı oranı
click_users = (user_stats["n_clicks"] > 0).mean()
print(f"En az 1 tıklama yapan kullanıcı oranı: {click_users:.2%}")

# Ortalama kullanıcı başına session, click, order
avg_sessions = user_stats["n_sessions"].mean()
avg_clicks   = user_stats["n_clicks"].mean()
avg_orders   = user_stats["n_orders"].mean()

print(f"Kullanıcı başına ortalama session sayısı: {avg_sessions:.2f}")
print(f"Kullanıcı başına ortalama click sayısı: {avg_clicks:.2f}")
print(f"Kullanıcı başına ortalama order sayısı: {avg_orders:.2f}")
````

````python
import numpy as np
# (1) Satışa göre indirim oranı
price_data["discount_rate_vs_selling"] = np.where(
    price_data["selling_price"].notna(),
    (price_data["selling_price"] - price_data["discounted_price"]) / price_data["selling_price"],
    np.nan
)

# (2) Orijinale göre indirim oranı
price_data["discount_rate_vs_original"] = np.where(
    price_data["original_price"].notna(),
    (price_data["original_price"] - price_data["discounted_price"]) / price_data["original_price"],
    np.nan
)
````

### 2.   görsel review oranını cıkart


````python
# Görsel yorum oranı için yeni kolon
price_data["visual_review_ratio"] = price_data["content_review_wth_media_count"] / price_data["content_rate_count"]

# 0’a bölme hatalarını önlemek için sonsuz/NaN değerleri 0 yapalım
price_data["visual_review_ratio"] = price_data["visual_review_ratio"].fillna(0)
````

### 1.   ctr hesapla


````python
# CTR hesapla
search_log["ctr"] = search_log["total_search_click"] / search_log["total_search_impression"]
````

````python
import numpy as np

search_log["ctr"] = np.where(
    search_log["total_search_impression"] > 0,
    search_log["total_search_click"] / search_log["total_search_impression"],
    0
)
````

### 2.   popülerlik skoru hesapla


````python
# Normalize edilmiş popülerlik skoru
total_impressions = search_log["total_search_impression"].sum()
search_log["popularity_score"] = search_log["total_search_impression"] / total_impressions
````

### 3.   kullanılmayacak kolonları drop et


````python
search_log.head()
````

````python
search_log.drop(columns=["total_search_impression", "total_search_click"], inplace=True)
````

### 4.   tekilleştir

````python
# content_id_hashed bazında tekilleştir
search_log = (
    search_log
    .groupby("content_id_hashed", as_index=False)
    .agg({
        "ctr": "mean",              # ortalama CTR
        "popularity_score": "sum",  # popülerlik skorlarını toplayabilirsin
        "date": "min"               # en eski tarihi al (ya da "max" yapabilirsin)
    })
)
````

boşları dolduralım

````python
# 1) Eksik bayrakları
for c in ["ctr", "popularity_score", "date"]:
    test_raw[f"{c}_was_null"] = test_raw[c].isna().astype(np.int8)

# 2) Sayısallar: ctr & popularity_score -> ürün (content_id_hashed) medyanı
for c in ["ctr", "popularity_score"]:
    grp_med = test_raw.groupby("content_id_hashed")[c].transform("median")
    test_raw[c] = test_raw[c].fillna(grp_med) \
                 .fillna(test_raw[c].median())  # hâlâ NaN ise global medyan

# 3) Tarih: medyan tarih ile doldur (datetime’a çevirip int üzerinden medyan)
test_raw["date"] = pd.to_datetime(test_raw["date"], errors="coerce")

# content_id_hashed bazında medyan tarih (ns int -> geri datetime)
date_int = test_raw["date"].astype("int64")  # NaT -> NaN kalır
grp_med_date_int = date_int.groupby(test_raw["content_id_hashed"]).transform("median")
grp_med_date = pd.to_datetime(grp_med_date_int)

# doldur: önce ürün medyanı, sonra global medyan
global_med_date = pd.to_datetime(date_int.median())  # tüm set medyan tarihi
test_raw["date"] = test_raw["date"].fillna(grp_med_date).fillna(global_med_date)
````

````python
# 1) Eksik bayrakları
for c in ["ctr", "popularity_score", "date"]:
    train_raw[f"{c}_was_null"] = train_raw[c].isna().astype(np.int8)

# 2) Sayısallar: ctr & popularity_score -> ürün (content_id_hashed) medyanı
for c in ["ctr", "popularity_score"]:
    grp_med = train_raw.groupby("content_id_hashed")[c].transform("median")
    train_raw[c] = train_raw[c].fillna(grp_med) \
                 .fillna(train_raw[c].median())  # hâlâ NaN ise global medyan

# 3) Tarih: medyan tarih ile doldur (datetime’a çevirip int üzerinden medyan)
train_raw["date"] = pd.to_datetime(train_raw["date"], errors="coerce")

# content_id_hashed bazında medyan tarih (ns int -> geri datetime)
date_int = train_raw["date"].astype("int64")  # NaT -> NaN kalır
grp_med_date_int = date_int.groupby(train_raw["content_id_hashed"]).transform("median")
grp_med_date = pd.to_datetime(grp_med_date_int)

# doldur: önce ürün medyanı, sonra global medyan
global_med_date = pd.to_datetime(date_int.median())  # tüm set medyan tarihi
train_raw["date"] = train_raw["date"].fillna(grp_med_date).fillna(global_med_date)
````

# **Model**

````python
import pandas as pd
import numpy as np
from sklearn.model_selection import GroupKFold
from sklearn.metrics import roc_auc_score
from lightgbm import LGBMClassifier

SEED = 42
````

````python
def session_auc(df, session_col, y_true_col, y_score_col):
    scores = []
    for sid, g in df.groupby(session_col):
        y = g[y_true_col].values
        if (y.sum() == 0) or (y.sum() == len(y)):
            continue
        s = roc_auc_score(y, g[y_score_col].values)
        scores.append(s)
    return float(np.mean(scores)) if scores else np.nan
````

````python
def prep(df, is_train=True):
    out = df.copy()

    # Tarih özellikleri
    if 'ts_hour' in out.columns:
        out['ts_hour_hour'] = out['ts_hour'].dt.hour.astype('int16')
        out['ts_hour_dow']  = out['ts_hour'].dt.dayofweek.astype('int8')
    else:
        out['ts_hour_hour'] = 0
        out['ts_hour_dow']  = 0


    # Sayısal kolonlar =
    num_cols = [
        'total_attribute_option_count',
        'merchant_count',
        'filterable_label_count',
        'original_price',
        'content_review_count',
        'content_rate_count',
        'content_rate_avg',
        'discount_rate_vs_selling',
        'discount_rate_vs_original',
        'visual_review_ratio',
        'ctr',
        'popularity_score',
        'total_click',
        'total_cart',
        'total_fav',
        'total_order',
        'term_ctr',
        'user_tenure_in_days',
        'user_search_imp_sum',
        'user_search_clk_sum',
        'user_search_ctr',
        'cv_tag_count',
        'product_age_days',
        'leaf_category_click_count', 'total_cart_mean', 'total_fav_mean',
        'total_order_mean', 'u_engagement', 'u_click_ratio', 'u_conv_rate',
        'u_cart2order','total_search_impression','total_search_click','content_ctr'
    ]


    num_cols = [c for c in num_cols if c in out.columns]

    # Kategorikler
    cat_cols = [
        'level1_category_name',
        'user_gender'
    ]
    cat_cols = [c for c in cat_cols if c in out.columns]
    for c in cat_cols:
        out[c] = out[c].astype('category')

    keep = num_cols + cat_cols + ['session_id','content_id_hashed']
    if is_train:
        targets = [c for c in ['clicked','ordered'] if c in out.columns]
        keep += targets
    return out[keep], num_cols, cat_cols

train_df, NUM_COLS, CAT_COLS = prep(train_raw, is_train=True)
````

````python
gkf = GroupKFold(n_splits=5)
groups = train_df['session_id'].values
FEATURES = NUM_COLS + CAT_COLS
````

````python
params_base = dict(
    n_estimators=1500,
    learning_rate=0.03,
    max_depth=-1,
    num_leaves=63,
    subsample=0.8,
    colsample_bytree=0.8,
    reg_alpha=1.0,
    reg_lambda=2.0,
    random_state=SEED,
    n_jobs=-1,
    device='gpu',
    gpu_platform_id=0,
    gpu_device_id=0
)

def fit_lgbm(train_df, target_col):
    oof_pred = np.zeros(len(train_df), dtype=float)
    models = []

    for fold, (tr_idx, va_idx) in enumerate(gkf.split(train_df, groups=groups)):
        tr, va = train_df.iloc[tr_idx], train_df.iloc[va_idx]

        clf = LGBMClassifier(**params_base)
        clf.fit(
            tr[FEATURES], tr[target_col],
            eval_set=[(va[FEATURES], va[target_col])],
            eval_metric='auc',
            categorical_feature=CAT_COLS
        )
        models.append(clf)
        oof_pred[va_idx] = clf.predict_proba(va[FEATURES])[:,1]

        va_tmp = va[['session_id', target_col]].copy()
        va_tmp['score'] = oof_pred[va_idx]
        fold_auc = session_auc(va_tmp, 'session_id', target_col, 'score')
        print(f"[{target_col}] Fold {fold} session-AUC: {fold_auc:.4f}")

    oof_df = train_df[['session_id', target_col]].copy()
    oof_df['score'] = oof_pred
    auc_final = session_auc(oof_df, 'session_id', target_col, 'score')
    print(f"[{target_col}] CV session-AUC: {auc_final:.4f}")
    return models, oof_pred, auc_final

click_models, click_oof, click_auc = fit_lgbm(train_df, 'clicked')
order_models, order_oof, order_auc = fit_lgbm(train_df, 'ordered')
````

````python
test_df, _, _ = prep(test_raw, is_train=False)

def predict_avg(models, X):
    preds = [m.predict_proba(X)[:,1] for m in models]
    return np.mean(preds, axis=0)

test_df['p_click'] = predict_avg(click_models, test_df[FEATURES])
test_df['p_order'] = predict_avg(order_models, test_df[FEATURES])

W_ORDER, W_CLICK = 0.7, 0.3
test_df['final_score'] = W_ORDER*test_df['p_order'] + W_CLICK*test_df['p_click']
````

````python
def make_submission(df, session_col='session_id', item_col='content_id_hashed', score_col='final_score'):
    rows = []
    for sid, g in df.groupby(session_col):
        g = g.sort_values(score_col, ascending=False)
        rows.append((sid, " ".join(g[item_col].astype(str).tolist())))
    return pd.DataFrame(rows, columns=['session_id','prediction'])

submission = make_submission(test_df)
submission.to_csv("submission.csv", index=False)
submission.head()
````

````python
# =========================
# LightGBM Feature Importance (avg over folds)
# =========================
def get_lgb_importance(lgb_models, feature_names, importance_type="gain"):
    """
    lgb_models: LGBMClassifier listesi (fold başına bir model)
    feature_names: FEATURES listesi
    importance_type: 'gain' veya 'split' (booster_ için)
    """
    imps = []
    for m in lgb_models:
        if hasattr(m, "feature_importances_"):
            # sklearn API
            imps.append(m.feature_importances_)
        elif hasattr(m, "booster_"):
            # native booster
            imps.append(m.booster_.feature_importance(importance_type=importance_type))
        else:
            # nadiren: native model objesi
            imps.append(m.feature_importance(importance_type=importance_type))

    imps = np.vstack(imps)
    mean_imp = imps.mean(axis=0)
    std_imp  = imps.std(axis=0)

    df = (pd.DataFrame({
            "feature": feature_names,
            "lgb_importance_mean": mean_imp,
            "lgb_importance_std": std_imp
          })
          .sort_values("lgb_importance_mean", ascending=False)
          .reset_index(drop=True))
    df["lgb_importance_rank"] = df["lgb_importance_mean"].rank(ascending=False, method="dense").astype(int)
    return df

# Click modeli için
click_importance = get_lgb_importance(click_models, FEATURES, importance_type="gain")
print("\nTop 10 Önemli Özellikler (Click - LightGBM):")
print(click_importance.head(44)[["feature","lgb_importance_mean"]])

# Order modeli için
order_importance = get_lgb_importance(order_models, FEATURES, importance_type="gain")
print("\nTop 10 Önemli Özellikler (Order - LightGBM):")
print(order_importance.head(44)[["feature","lgb_importance_mean"]])

# (İstersen CSV'ye yaz)
# click_importance.to_csv("fi_click_lgb.csv", index=False)
# order_importance.to_csv("fi_order_lgb.csv", index=False)
````
